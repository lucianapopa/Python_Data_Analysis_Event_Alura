```python
df = df.rename(columns={
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
})

display(df.head())

# As part of the practice, I've changed the column names from English to Brazilian Portuguese.

# If I decide to rename columns again, the previous line must be previously runned.
```

```python
df["senioridade"].value_counts()

# Verifying each category frequency within the 'experience level' column.

# It is possible to see that the majority of analysed employees are seniors.
```

```python
df["contrato"].value_counts()

# Employment contrat frequency: full-time, contract, part-time, freelancer.

# The majority of contracts are full-time.
```

```python
df["remoto"].value_counts()

# 0 = presencial/ on-site
# 50 = remoto/ remote
# 100 = hibridas/ hybrid

# The majority of employees work on-site.
```

```python
df["tamanho_empresa"].value_counts()

# The majority of companies are medium-sized.
```

```python
# Renaming the categories within the columns replacing abbreviations for full terms for better understanding.
```

```python
senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Especialista'
}

df['senioridade'] = df['senioridade'].replace(senioridade)

display(df.head())
```

```python
contrato = {
    'FT': 'Full-time',
    'CT': 'Contract',
    'PT': 'Part-time',
    'FL': 'Freelancer'
}

df['contrato'] = df['contrato'].replace(contrato)

display(df.head())
```

```python
tamanho_empresa = {
    'S': 'Startup',
    'M': 'Médio',
    'L': 'Grande'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)

display(df.head())
```

```python
remoto = {
    0: 'Presencial',
    50: 'Remoto',
    100: 'Híbrido'
}

df['remoto'] = df['remoto'].replace(remoto)

display(df.head())
```
